```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
X  = np.load("X.npy")
Y = np.load("Y.npy")
print(X.shape)
print(Y.shape)
print(np.shape(X[1,:]))
```

```python
p = 0
idx = []
idx.append(0)
for i in range(0,len(Y)):
    if np.argmax(Y[i]) != p:
        idx.append(i)
        p = p+1

print(idx)
```

```python
fig=plt.figure(figsize=(20, 20))
columns = 10
rows = 1
for i in range(1, columns*rows +1):
    img = X[idx[i-1]]
    fig.add_subplot(rows, columns, i)
    plt.imshow(img)
    print(Y[idx[i-1]])
plt.show()
```
